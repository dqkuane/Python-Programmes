```python
class SudokuSolver():
    
    def __init__(self, board):
        self.unsolved = board
        self.board = board
        self.solved = None
    
    
    def find_empty(self):
        board = self.board
        for i in range(len(board)):
            for j in range(len(board[0])):
                if board[i][j] == 0:
                    return (i, j)
        
        #if there is nothing we return None
        return None
    
    def valid(self, board, num, pos):
        
        for i in range(len(board[0])):
            #checking each column in the row to see if the same number is in the row
            #also not i to make sure we don't disaqualify based on
            #the column of the value we just put in
            #iterating across
            if board[pos[0]][i] == num and pos[1] != i:
                return False
            
            #same for the rows in each column
            #iterating down 
            if board[pos[1]][i] == num and pos[0] != i:
                return False


        #Checking Box
        x_coord = pos[1] // 3
        y_coord = pos[0] // 3
        
        #box x and y result in 0,1 or 2 bc it's integer division (e.g. 8//3 = 2)

        for i in range(y_coord * 3, y_coord*3 + 3):
            for j in range(x_coord * 3, x_coord*3 + 3):
                if board[i][j] == num and (i, j) != pos:
                    return False
        #so to get the index position
            #we do x*3 to get the index of the top left of the box
            #and, do x*3 +3, for x coordinates, to get the top right
                #or for y coordinates, the bottom left
            #and we iterate over this little band of three
                #to check if the number is in the columns (using j)
                #for each row (i)
    
        return True
    
    def print_board(self, solved_board = None):
        if solved_board == None:
            board = self.unsolved
        else:
            board = self.solved
            
        for i in range(len(board)):
            if i % 3 == 0 and i != 0:
                print('                            ')

            for j in range(len(board[0])):
                if j % 3 == 0 and j != 0:
                    print('   ', end = '')

                if j == 8:
                    print(board[i][j])

                else:
                    print(str(board[i][j]) + " ", end = '')
```

```python
class Solving(SudokuSolver):
    
    def __init__(self, board):
        super().__init__(board)
    
    def solve(self):

        board = self.board
        #recursive strategy
        empty = find_empty(board)
        if not empty:
            return True
        else:
            row, col = empty

        for i in range(1,10):
            if valid(board, num = i, pos = (row, col)):
                board[row][col] = i
                
                #now we've entered the number, we try to solve w this
                if self.solve():
                    return True

                board[row][col] = 0
        
        #if we loop through all numbers and none are valid we return False
        #so solve can't be True
        #and thus we change board[row][col] back to 0 
        
        self.solved = board
        return False
```
